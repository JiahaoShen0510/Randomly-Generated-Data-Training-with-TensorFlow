```python
import numpy as np
import tensorflow as tf
```

```python
# create randomly generated dataset for algorithm training
```

```python
observation=10000
xs=np.random.uniform(low=-10,high=10,size=(observation,1))
zs=np.random.uniform(low=-10,high=10,size=(observation,1))
inputs=np.column_stack((xs,zs))

noises=np.random.uniform(low=-1,high=1,size=(observation,1))
```

```python
print(inputs.shape,noises.shape)
```

```python
target=634*xs-897*zs-13+noises
```

```python
np.savez("Random Data",inputs=inputs,targets=target)
```

```python
training_data=np.load("Random Data.npz")
```

```python
input_size=2
output_size=1
```

```python
# L2 approach with Tensorflow
```

```python
model_01=tf.keras.Sequential([tf.keras.layers.Dense(output_size,
                          kernel_initializer=tf.random_uniform_initializer(-0.1,0.1), 
                          bias_initializer=tf.random_uniform_initializer(-0.1,0.1)
                                                )
                          ])
optimizer_01=tf.keras.optimizers.SGD(learning_rate=0.01)
model_01.compile(optimizer=optimizer_01,loss="mean_squared_error")
model_01.fit(training_data["inputs"],training_data["targets"],epochs=100,verbose=1)
```

```python
weights=model_01.layers[0].get_weights()[0]
bias=model_01.layers[0].get_weights()[1]
```

```python
print(weights)
print(bias)
```

```python
# L1 approach with Tensorflow
```

```python
model_02=tf.keras.Sequential([tf.keras.layers.Dense(output_size,
                                                    kernel_initializer=tf.random_uniform_initializer (-0.1,0.1),
                                                    bias_initializer=tf.random_uniform_initializer (-0.1,0.1)
                                                   )
                             ])
optimizer_02=tf.keras.optimizers.SGD(learning_rate=0.01)     
model_02.compile(optimizer=optimizer_02, loss="mean_absolute_error")
model_02.fit(training_data["inputs"],training_data["targets"],epochs=100,verbose=1)
```

```python
weights=model_02.layers[0].get_weights()[0]
bias=model_02.layers[0].get_weights()[1]
```

```python
print(weights)
print(bias)
```
